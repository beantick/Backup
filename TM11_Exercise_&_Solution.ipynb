**20. Buatlah kolom baru yang isinya adalah income category (low, medium, high) berdasarkan variabel penghasilan tahunan**

```python
df['annual_income'].describe()
```

```python
# Insert your code here
income_categories = ['low', 'medium', 'high']

# Bagi values dari kolom annual_income 
# dengan interval yang sama menggunakan 
# pd.qcut()
categories = pd.qcut(df['annual_income'], 
q=3, labels=income_categories)

# Buat kolom baru bernama income_category
df['income_category'] = categories

# Tampilkan hasil
df[['annual_income', 'income_category']]
```

```python
# Tampilkan data yang income_category nya high
# Insert your code here
df[df['income_category'] == 'high'
].sort_values(by='annual_income')
```

**23. Buatlah DataFrame untuk menampilkan nilai rata-rata dari penghasilan tahunan, resensi, serta jumlah pembelian pada toko dan online berdasarkan kelompok pada kolom level pendidikan**

```python
# Insert your code here
mean_aggregation = df.groupby(
    by='educational_level').mean(
    )[['annual_income', 'recency', 
    'store_purchases', 'online_purchases']]
    
mean_aggregation
```

```python
# Reset index
mean_aggregation.reset_index()
```

**24. Buatlah pivot table untuk menampilkan data jumlah pembelian dan rata-rata resensi berdasarkan kelompok pada kolom level pendidikan dan status pernikahan**

```python
# Insert your code here
df_pivot = pd.pivot_table(df,
    values=['recency', 'sum_purchases'],
    index=['marital_status'],
    columns=['educational_level'],
    aggfunc={'recency': 'mean', 
            'sum_purchases': 'sum'},
    fill_value=0
    ).reset_index()

df_pivot
```
